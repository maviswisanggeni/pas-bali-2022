#### Plotting raster in Python.  

First, let's explore some information from the geotiff file such as the projection info.

```python
raster_wkt = dtm.GetProjection()
spatial_ref = osr.SpatialReference()
spatial_ref.ImportFromWkt(raster_wkt)
spatial_ref.ExportToProj4()
```

Now, plotting with matplotlib.

```python
import matplotlib.pyplot as plt
plDTM = dtm.ReadAsArray()
plt.imshow(plDTM, cmap=plt.cm.terrain)
plt.colorbar()
plt.show()
```
