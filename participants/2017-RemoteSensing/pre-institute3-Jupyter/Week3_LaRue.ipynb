```python
 #open TIFF file from current directory using gdal.Open
SJER_DTM_17 = gdal.Open('NEON_D17_SJER_DP3_252000_4109000_DTM.tif')
```

```python
SJER_DTM_17.RasterXSize
```

2. Plot the TIFF 

```python
#extract band 1 and convert to an array in order to plot
bnd1 = SJER_DTM_17.GetRasterBand(1)
bnd1Array = bnd1.ReadAsArray() 
```

```python
#import to plot 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
%matplotlib inline
plt.style.use("bmh")

# create plot dimensions
fig = plt.figure(figsize=(10,8))

#plot image
ax = fig.add_subplot(111)
imgplot = ax.imshow(bnd1Array)
plt.show()
```
